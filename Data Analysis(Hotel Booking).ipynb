## Data Analysis and Visualizations

```python
df.head(2)
```

```python
canceled_pres=df.is_canceled.value_counts(normalize=True)
print(canceled_pres)
plt.figure(figsize=(5,4))
plt.title("Reservation status count")
plt.bar(["Not canceled","canceled"],df.is_canceled.value_counts(),width=0.7)
plt.show()
```

```python
plt.figure(figsize=(8,4))
ax1=sns.countplot(x="hotel",hue="is_canceled",data=df,edgecolor="k")
plt.title("Reservation status of different hotel")
plt.xlabel("hotel")
plt.ylabel("number of reservation")
plt.legend(["Not canceled","canceled"])
plt.show()
```

```python
resort_hotel=df[df.hotel=="Resort Hotel"]
resort_hotel["is_canceled"].value_counts(normalize=True)
```

```python
city_hotel=df[df.hotel=="City Hotel"]
city_hotel["is_canceled"].value_counts(normalize=True)
```

```python
resort_hotel=resort_hotel.groupby("reservation_status_date")[["adr"]].mean()
city_hotel=city_hotel.groupby("reservation_status_date")[["adr"]].mean()
```

```python
plt.figure(figsize=(10,4))
plt.title("Avarage Daily Rate in City and Resort Hotel")
plt.plot(resort_hotel.index,resort_hotel["adr"],label="Resort Hotel")
plt.plot(city_hotel.index,city_hotel["adr"],label="City Hotel")
plt.legend()
plt.show()
```

```python
df["month"]=df["reservation_status_date"].dt.month
plt.figure(figsize=(10,4))
sns.countplot(x="month",hue="is_canceled",data=df)
plt.legend(["Not Cencelled","Cencelled"])
plt.title("Reservation Status Per Month")
plt.ylabel("Number of Reservation")
plt.show()
```

```python
plt.figure(figsize=(10,5))
plt.title("ADR per month",size=20,color="g")
sns.barplot("month","adr",data=df[df.is_canceled==1].groupby("month")[["adr"]].sum().reset_index())
plt.show()
```

```python
canceled_data=df[df.is_canceled==1]
top_10_country=canceled_data["country"].value_counts()[:10]
plt.figure(figsize=(8,8))
plt.pie(x=top_10_country.values,autopct="%0.2f",labels=top_10_country.index)
plt.title("Top 10 Country with Reservation Canceled")
plt.show()
```

```python
df.market_segment.value_counts()
```

```python
df.market_segment.value_counts(normalize=True)
```

```python
canceled_data.market_segment.value_counts(normalize=True)
```

```python
Non_canceled_data=df[df.is_canceled==0]
Non_canceled_data.market_segment.value_counts(normalize=True)
```

```python
canceled_df_adr=canceled_data.groupby("reservation_status_date")[["adr"]].mean()
canceled_df_adr.reset_index(inplace=True)
canceled_df_adr.sort_values("reservation_status_date",inplace=True)

Non_canceled_df_adr=Non_canceled_data.groupby("reservation_status_date")[["adr"]].mean()
Non_canceled_df_adr.reset_index(inplace=True)
Non_canceled_df_adr.sort_values("reservation_status_date",inplace=True)

plt.figure(figsize=(15,5))
plt.title("Avarage Daily Rate")
plt.plot(canceled_df_adr["reservation_status_date"],canceled_df_adr["adr"],label="Canceled")
plt.plot(Non_canceled_df_adr["reservation_status_date"],Non_canceled_df_adr["adr"],label="Non Canceled")
plt.legend()
plt.show()
```

```python
canceled_df_adr=canceled_df_adr[(canceled_df_adr.reservation_status_date>"2016") & (canceled_df_adr.reservation_status_date<"2017-09")]
Non_canceled_df_adr=Non_canceled_df_adr[(Non_canceled_df_adr.reservation_status_date>"2016") & (Non_canceled_df_adr.reservation_status_date<"2017-09")]
```

```python
plt.figure(figsize=(20,5))
plt.title("Avarage Daily Rate",c="g",size=20)
plt.plot(canceled_df_adr["reservation_status_date"],canceled_df_adr["adr"],label="Canceled")
plt.plot(Non_canceled_df_adr["reservation_status_date"],Non_canceled_df_adr["adr"],label="Non Canceled")
plt.legend(fontsize=15)
plt.show()
```
